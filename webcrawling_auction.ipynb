```python
## 옥션에서 bag 검색해서 가격, 제품명, 판매자, (할인율) 크롤링하기.
## 없는 항목 처리
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "http://browse.auction.co.kr/search?keyword=bag&itemno=&nickname=&frm=hometab&dom=auction&isSuggestion=No&retry=&Fwk=bag&acode=SRP_SU_0100&arraycategory=&encKeyword=bag"
page = urlopen(url)
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
info_all = soup.find_all('div', class_="section--itemcard_info")
```

```python
num_list = []
name_list = []
cnt = 1
for list in info_all:
    name = list.find('span', class_="text--title")
    name_list.append(name.text)
    num_list.append(cnt)
    cnt = cnt+1
print(cnt)
```

```python
price_list = []
for list in info_all:
    price = list.find('strong', class_="text--price_seller")
    price_list.append(price.text)  
```

```python
seller_list = []
for list in info_all:
    seller_info = list.find('div', class_="section--itemcard_info_shop")
    try :                                  # 예외처리구문
        seller = seller_info.find('span', class_="img")       
        seller_list.append(seller.text)
    except :
        try :
            seller = seller_info.find('span', class_="text")
            seller_list.append(seller.text)
        except :
            seller_list.append('정보없음') 
```

```python
rate_list = []
for list in info_all:
    try:
        rate = list.find('strong', class_="text--discount")
        rate = rate.text + '%'
        rate_list.append(rate)
    except:
        rate_list.append("-")
```

```python
import pandas as pd
```

```python
fnum_list= pd.Series(num_list)
fname_list= pd.Series(name_list)
fprice_list= pd.Series(price_list)
frate_list = pd.Series(rate_list)
fseller_list= pd.Series(seller_list)
dat_ori = {'num':fnum_list, '제품명':fname_list, '가격':fprice_list, '할인율':frate_list, '판매자':fseller_list}
dat = pd.DataFrame(dat_ori, columns = ['num', '제품명', '가격', '할인율', '판매자'])
dat
```

```python
dat.to_csv("auction_infor.csv", index=False, encoding="EUCKR")
```
