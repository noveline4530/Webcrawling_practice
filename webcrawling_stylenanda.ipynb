```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
url = "https://www.stylenanda.com/?cafe_mkt=google_kw&gclid=Cj0KCQiAkMDiBRDNARIsACKP1FGMh7eJ_o1u1gPan4Ik2Hz185H_Nff6mvSLeU6PqZPFOILWH08GKzoaAl-fEALw_wcB#none"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
```

```python
item_table = soup.find_all('div', class_='table')
item_infor = soup.find_all('div', class_='information')
```

```python
item_list = []
for com in item_table:
    name = com.find('span')
    item_list.append(name.text)
len(item_list)
```

```python
for com in item_infor:
    try:
        name = com.find('span')
        item_list.append(name.text)
    except:
        continue
item_list
```

```python
len(item_list)
```

```python
price_list = []
for com in item_table:
    raw_price = com.find('p', class_='price')
    price = raw_price.text.split(' ')[0]    # 가격만 정확히 파싱
    price_list.append(price)
price_list
```

```python
len(price_list)
```

```python
for com in item_infor:
    try:
        price = com.find('p', class_='price')
        if len(price.text.split()) == 2:
            price_list.append(price.text.split(' ')[0])            
        else:
            price_list.append(price.text.strip())
    except:
        continue
price_list
```

```python
len(price_list)
```

```python
import pandas as pd
```

```python
item = pd.Series(item_list)
price = pd.Series(price_list)
dat_ori = {'상품명':item, '가격':price}
dat = pd.DataFrame(dat_ori, columns = ['상품명', '가격'])
dat
```

```python
dat.to_csv("stylenanda_infor.csv", index = False, encoding="EUCKR")
```
